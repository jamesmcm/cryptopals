```python
f=open("8.txt","r")
cps=[]
for line in f:
    cps.append(line.strip())
f.close()
```

```python
s="longteststrings1"
print len(s)
```

```python
lenblock=4
for i in xrange(len(s)/lenblock):
   for j in xrange(i+1,(len(s)/lenblock)):
        print s[i*lenblock:(i+1)*lenblock] + s[j*lenblock:(j+1)*lenblock]
 
```

```python
lenblock=32
simd={}
for z in xrange(len(cps)):
    c=cps[z]
    count=0
    for i in xrange(len(c)/lenblock):
       for j in xrange(i+1,(len(c)/lenblock)):
            if c[i*lenblock:(i+1)*lenblock] == c[j*lenblock:(j+1)*lenblock]:
                count+=1
            simd[z]=count
                
```

```python
print simd
```

```python
import operator
sorted_x = sorted(simd.items(), key=operator.itemgetter(1), reverse=True)
print sorted_x
```
